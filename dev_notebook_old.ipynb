```python
class Table:
    def __init__(self, capacity: int = 4, seats:list = None):
        self.capacity = capacity # integer
        self.seats = [Seat() for _ in range(capacity)] # list of Seat objects ***
        
    def has_free_spot(self) -> bool: # returns a boolean (True if a spot is available)
        if len(self.seats) <  self.capacity:
            self.has_free_spot = True
        else:
            self.has_free_spot = False
        return(self.has_free_spot)

    def assign_seat(self, seat_name): # places someone at the table
        if seat in self.seats:
            if seat.is_free:
                seat.set_occupant(seat_name) # we add 1 seat to the table ***
                return()
            raise ValueError('No free seats at this table') # Exception when the program execution is pointless***
    
    def left_capacity(self)-> int: # returns an integer
        count_free_seat = 0
        for seat in self.seats:
            if seat.is_free==True:
                count_free_seat+=1
        return(count_free_seat)
                
```

```python
import random
# from table.py import list_of_tables

class Openspace:
    def __init__(self, tables : list, number_of_tables:int=6):
        self.tables = tables # list of Table (import Table from table.py).
        self.number_of_tables = number_of_tables # 

    def organize(self,names): # randomly assign people to Seat objects in the different Table objects.
        random.shuffle(names)
        all_seats=[seat for table in self.tables for seat in table.seats]

        #if len(names)>len(all-seats):
        #    raise ValueError('Not enough seats for everyone')
    
        for seat, name in zip(all_seats,names):
            seat.occupant=name

    def display(self): # display the different tables and there occupants in a nice and readable way.
        for i, table in enumerate(self.tables, start=1):
            print(f'\nTable {i}:')
            for seat in table.seats:
                occupant = seat.occupant if seat.occupant else 'Free'
                print(f' - {occupant}')
        
    def store(filename): # store the repartition in an file.
        data={}
        for i, table in enumerate(self.tables, start =1):
            data[f'Table{i}']=[seat.occupant if seat.occupant else None for seat in table.seats]

        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(data,f,indent=4, ensure_ascii=False)
        print(f"\nSeating plan saved to '{filename}'")
```
